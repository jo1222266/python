```python
  ##################

  #프로그램명: 성적 관리 프로그램

  #작성자: 소프트웨어학과/조예지

  #작성일: 2025/04/15

  #프로그램 설명: 성적관리 프로그램
  # 1. 5명의 학생에 대해서 학번, 이름, 영어, C언어, 파이썬 점수를 입력받습니다.
  # 2. 총점, 평균, 학점, 등수를 계산하여 출력합니다
  # 3. 삽입, 삭제, 탐색, 정렬, 80점 이상 학생 수 카운트 기능이 있습니다.

  ###################

from typing import List, Tuple

st_info: List[Tuple[str, str, int, int, int, int, float, str, int]] = []

# 학생 정보 입력 함수
def get_info():
    id, name, eng, c, python = input("학번, 이름, 영어, C-언어, 파이썬 점수를 입력하세요 (공백으로 구분): ").split()
    eng, c, python = int(eng), int(c), int(python)
    total = get_total(eng, c, python)
    avg = get_avg(total)
    grade = get_grade(avg)
    st_info.append((id, name, eng, c, python, total, avg, grade, 0))

# 총점 계산 함수
def get_total(eng, c, python):
    return eng + c + python

# 평균 계산 함수
def get_avg(total):
    return round(total / 3, 2)

# 학점 계산 함수
def get_grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 80:
        return 'B'
    elif avg >= 70:
        return 'C'
    elif avg >= 60:
        return 'D'
    else:
        return 'F'

# 등수 계산 함수
def calculate_ranks():
    global st_info
    st_info.sort(key=lambda x: x[5], reverse=True)
    for rank, student in enumerate(st_info, start=1):
        st_info[rank - 1] = student[:8] + (rank,)

# 성적 출력 함수
def print_scores():
    print("\n성적 관리 프로그램")
    print("=" * 80)
    print("학번\t\t이름\t영어\tC-언어\t파이썬\t총점\t평균\t학점\t등수")
    print("=" * 80)
    for student in st_info:
        print(f"{student[0]}\t{student[1]}\t{student[2]}\t{student[3]}\t{student[4]}\t"
              f"{student[5]}\t{student[6]}\t{student[7]}\t{student[8]}")

# 삽입 함수
def insert_info():
    get_info()
    calculate_ranks()

# 학생 정보 삭제 함수
def delete_info(student_id):
    global st_info
    st_info = [student for student in st_info if student[0] != student_id]
    calculate_ranks()

# (학번) 탐색 함수
def search_by_id(student_id):
    return next((student for student in st_info if student[0] == student_id), None)

# (이름) 탐색 함수
def search_by_name(name):
    return [student for student in st_info if student[1] == name]

# 총점 정렬 함수
def sort_by_total():
    global st_info
    st_info.sort(key=lambda x: x[5], reverse=True)

# 80점 이상 학생 수 카운트 함수
def count_above_80():
    return len([student for student in st_info if student[6] >= 80])



for i in range(5):
    get_info()

calculate_ranks()
print_scores()


while True:
    print("\n---- 메뉴 ----")
    print("1. 학생 정보 삽입")
    print("2. 학생 정보 삭제")
    print("3. 학번으로  탐색")
    print("4. 이름으로  탐색")
    print("5. 총점 기준 정렬")
    print("6. 80점 이상 학생 수 카운트")
    print("7. 프로그램 종료")
    choice = input("원하는 번호를 선택하세요: ")

    if choice == "1":
        insert_info()
        print_scores()
    elif choice == "2":
        sid = input("삭제할 학생의 학번 입력: ")
        delete_info(sid)
        print_scores()
    elif choice == "3":
        sid = input("탐색할 학생의 학번 입력: ")
        result = search_by_id(sid)
        if result:
            print("찾은 학생 정보:", result)
        else:
            print("해당 학번 학생 정보가 없습니다.")
    elif choice == "4":
        name = input("탐색할 학생 이름 입력: ")
        results = search_by_name(name)
        if results:
            for student in results:
                print("찾은 학생 정보:", student)
        else:
            print("해당 이름의 학생 정보가 없습니다.")
    elif choice == "5":
        sort_by_total()
        print("정렬 완료")
        print_scores()
    elif choice == "6":
        count = count_above_80()
        print("80점 이상 학생 수:", count)
    elif choice == "7":
        print("프로그램 종료.")
        break
    else:
        print("다시 시도하세요.")
```
